```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = "./results/Model Loss Comparison.csv"  # Replace with the actual path to your CSV file
results_df = pd.read_csv(data_path)
results_df
```

```python
colors = ['blue', 'green', 'red', 'purple']

plt.figure(figsize=(10, 6))  
for i, col in enumerate(results_df.columns[1:]):
    plt.plot(results_df["Epoch"], results_df[col], label=col, color=colors[i])

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Comparison of Different Encoder/Decoder Architectures")

plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
